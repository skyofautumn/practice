# 13장-1. 과적합 피하기_초음파 광물 예측

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
#from keras import backend as K

import pandas as pd
import numpy as np
import tensorflow as tf

#seed값 설정
seed=0
np.random.seed(seed)
tf.random.set_seed(3)

#데이터 입력
df = pd.read_csv('../data/sonar.csv', header=None)

print(df.info())
print(df.head())
```

```python
dataset = df.values
X = dataset[:,0:60].astype(np.float32)
Y_obj = dataset[:,60]

#문자열 반환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
print(Y)
```

```python
#학습 셋과 테스트 셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3,random_state=seed)
#70%, 30%

#모델설정
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #60과 1의 값은 틀리면 안된다.
#3개이상은 softmax, 두개는 sigmoid쓴다.

#모델 컴파일
model.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])

# 모델 실행
model.fit(X_train, Y_train, epochs=130, batch_size=5)

#결과 출력
print("\n Test Accuracy : %.2f" % (model.evaluate(X_test,Y_test)[1]))
```

* 학습셋과 테스트셋을 나뉘 83%가 나옴
나누기전은 100% 과적합 나옴
