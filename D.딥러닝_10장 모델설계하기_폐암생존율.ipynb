* 폐암 수술 환자의 생존율 예측하기(1장에서 했음)

```python
#딥러닝을 구동하는데 필요한 케라스 함수 호출(tensor가 keras를 취함)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#필요한 라이브러리 불러옴
import numpy as np
import tensorflow as tf
```

```python
#시행할때마다 같은 결과를출력하기 위해 설정하는 부분
np.random.seed(3)
tf.random.set_seed(3)
```

```python
#준비된 수술 환자 데이터를 불러옴
Data_set = np.loadtxt("../data/ThoraricSurgery.csv",delimiter=",")
```

```python
#환자의 기록과 수술 결과를 x와 y로 구분하여 저장
X = Data_set[:,0:17]
Y = Data_set[:,17]
```

```python
#딥러닝 구조를 결정(모델을 설정하고 실행하는 부분)
model = Sequential() #순차적으로
model.add(Dense(30, input_dim=17, activation='relu')) 
#sigmoid 사그라드는 기울기때문에 relu로 씀
#input_dim은 맨첨에 딱 한번만 적어줌
#입력으로 17개 들어갑니다.(데이터에서 17개의 값을 받아 은닉층의 30개 노드로 보낸다는 뜻)
#30개 출력이 나옴
model.add(Dense(1,activation='sigmoid')) 
#마지막결과로 출력층노트는 1개 #S자 곡선(sigmoid)
```

```python
from IPython.display import Image 
Image("../data/0222.jpg")
```

```python
#딥러닝 실행
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
#mean_squared_error(평균제곱오차함수-실제값 예측값 오차 줄여나감 loss) 
#metrics 평가지표(accuracy 정확도)
#compile 컴퓨터가 알아먹게 0과 1로 바꿔줌
model.fit(X, Y, epochs=100, batch_size=10)
```
